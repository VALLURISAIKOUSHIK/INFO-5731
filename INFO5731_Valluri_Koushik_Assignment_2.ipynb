# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(2) Collect the top 1000 User Reviews of a movie recently in 2023 or 2024 (you can choose any movie) from IMDB. [If one movie doesn't have sufficient reviews, collect reviews of atleast 2 or 3 movies]

(3) Collect all the reviews of the top 1000 most popular software from G2 or Capterra.

(4) Collect the **abstracts** of the top 10000 research papers by using the query "machine learning", "data science", "artifical intelligence", or "information extraction" from Semantic Scholar.

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
'''In the below code, I am scraping the top 1000 reviews for Oppenheimer Movie'''
#Importing Packages
from bs4 import BeautifulSoup
import pandas as pd
import requests
#Defined upper cap for review i.e. 1000
num_reviews=1000
#IMBD url to access Oppenheimer movie reviews page.
base_url=f"https://www.imdb.com/title/tt15398776/reviews"
#Defining a blank list where reviews will be stored afterscraping
movie_reviews=[]
#Indexing web page starting from zero untill we reach 1000 reviews
start=0
#Conditional check to verify if reviews length is 1000 or not
while len(movie_reviews) < num_reviews:
  #Requesting access to the URL
  html_text=requests.get(f"{base_url}?start={start}").text
  #parsing to lxml format
  soup=BeautifulSoup(html_text, 'lxml')
  #finding a specific class which contains reviews
  reviews=soup.find_all('div',class_='text show-more__control')
  #extracting only text from individual reviews and appending it to the above list
  for review in reviews:
    movie_reviews.append(review.text)
  #Incrementation for web page index
  start += 10
#prining the lenth of movie_reviews and first 6 elements of the list for verification
print(len(movie_reviews))
df=pd.DataFrame({"Reviews":movie_reviews})
df.head(10)
df.to_csv('movie_reviews.csv')
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
#tokens
#nltk.word_tokenize(movie_reviews3[0])
movie_reviews4=[]
for num in range(0,len(movie_reviews3)):
  movie_reviews4.append(nltk.word_tokenize(movie_reviews3[num]))
```

```python
# Remove stopwords
stop_words = set(stopwords.words('english'))
movie_reviews5=[]
for num in range(0,len(movie_reviews4)):
  movie_reviews5.append([word for word in movie_reviews4[num] if word not in stop_words])
```

```python
#Stemming
stemmer = PorterStemmer()
movie_reviews6=[]
for num in range(0,len(movie_reviews5)):
  movie_reviews6.append([stemmer.stem(word) for word in movie_reviews5[num]])
```

```python
# Lemmatization
#lemmatized_tokens = [lemmatizer.lemmatize(word) for word in stemmed_tokens]
lemmatizer = WordNetLemmatizer()
movie_reviews7=[]
for num in range(0,len(movie_reviews5)):
  movie_reviews7.append([lemmatizer.lemmatize(word) for word in movie_reviews5[num]])
```

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
import spacy
from collections import Counter
import nltk
nltk.download('averaged_perceptron_tagger')
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
#POS
for x in range(0,len(df["Cleaned Reviews"])):
  pos=nltk.pos_tag(df["Cleaned Reviews"][x].split(' '))
  print("Review "+str(x+1)+" : POS")
  print(pos)
  print("Review "+str(x+1)+" : POS_Count")
  print(Counter(tag for word, tag in pos))
  print("\n")
```

```python
#Dependency Parsing
for x in range(0,len(df["Cleaned Reviews"])):
  print("Review "+str(x+1)+" : Dependency Parsing")
  for t in nlp(df['Cleaned Reviews'][x]):
    print(t.text,'=>',t.dep_,'=>',t.head.text)
  print("\n")
#As you can see in the below output dependency parsing shows the dependence of words on each other.
```
